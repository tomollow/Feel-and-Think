```python
# 
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt

# 
X, y = make_blobs(n_samples=150, centers=3, cluster_std=1.0, random_state=42)

# 
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='viridis', marker='o', edgecolor='black', s=50)
plt.title("Generated Data")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

```python
# 
from scipy.cluster.hierarchy import linkage, dendrogram

# 
linked = linkage(X, method='ward')

# 
plt.figure(figsize=(10, 5))
dendrogram(linked, orientation='top')
plt.title("Hierarchical Clustering Dendrogram")
plt.xlabel("Sample Index")
plt.ylabel("Distance")
plt.show()
```

```python
# 
from scipy.cluster.hierarchy import fcluster

# 
clusters = fcluster(linked, 3, criterion='maxclust')

# 
plt.scatter(X[:, 0], X[:, 1], c=clusters, cmap='viridis', marker='o', edgecolor='black', s=50)
plt.title("Hierarchical Clustering Result")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

```python
# 
X_large_variance, y_large_variance = make_blobs(n_samples=150, centers=3, cluster_std=2.5, random_state=42)

# 
plt.scatter(X_large_variance[:, 0], X_large_variance[:, 1], c=y_large_variance, cmap='viridis', marker='o', edgecolor='black', s=50)
plt.title("Generated Data with Large Variance")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

```python
# 
linked_large_variance = linkage(X_large_variance, method='ward')

# 
plt.figure(figsize=(10, 5))
dendrogram(linked_large_variance, orientation='top')
plt.title("Hierarchical Clustering Dendrogram (Large Variance)")
plt.xlabel("Sample Index")
plt.ylabel("Distance")
plt.show()
```

```python
# 
clusters_large_variance = fcluster(linked_large_variance, 3, criterion='maxclust')

# 
plt.scatter(X_large_variance[:, 0], X_large_variance[:, 1], c=clusters_large_variance, cmap='viridis', marker='o', edgecolor='black', s=50)
plt.title("Hierarchical Clustering Result (Large Variance)")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

```python
# 
from scipy.cluster.hierarchy import fcluster

# 
clusters_large_variance = fcluster(linked_large_variance, 3, criterion='maxclust')

# 
plt.scatter(X_large_variance[:, 0], X_large_variance[:, 1], c=clusters_large_variance, cmap='viridis', marker='o', edgecolor='black', s=50)
plt.title("Hierarchical Clustering Result (Large Variance)")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```
