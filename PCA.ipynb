### 必要なライブラリのインポート

```python
# 必要なライブラリのインポート
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
```

### 乱数シードの設定

```python
# 乱数シードの設定
np.random.seed(42)
```

### 3次元データの生成

```python
# 3次元データの生成
mean = [0, 0, 0]
cov = [[1, 0.8, 0.5], [0.8, 1, 0.3], [0.5, 0.3, 1]]  # 共分散行列
X = np.random.multivariate_normal(mean, cov, 200)
```

### PCAの実行

```python
# PCAの実行
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)
```

### 元のデータのプロット

```python
# 元のデータのプロット
fig = plt.figure(figsize=(12, 6))
ax1 = fig.add_subplot(121, projection='3d')
ax1.scatter(X[:, 0], X[:, 1], X[:, 2], c='blue', marker='o', alpha=0.5)
ax1.set_title('Original Data')
ax1.set_xlabel('X1')
ax1.set_ylabel('X2')
ax1.set_zlabel('X3')
```

### 主成分分析後のデータのプロット

```python
# 主成分分析後のデータのプロット
ax2 = fig.add_subplot(122)
ax2.scatter(X_pca[:, 0], X_pca[:, 1], c='red', marker='o', alpha=0.5)
ax2.set_title('After PCA')
ax2.set_xlabel('Principal Component 1')
ax2.set_ylabel('Principal Component 2')

plt.show()
```
